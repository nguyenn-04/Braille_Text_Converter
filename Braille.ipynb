```python
import csv

# Hàm đọc dữ liệu từ file CSV
def read_braille_data(file_path):
    braille_dict = {}
    with open(file_path, mode='r') as file:
        reader = csv.reader(file)
        next(reader)  # Bỏ qua dòng tiêu đề
        for row in reader:
            braille_dict[row[0]] = row[1].split()
    return braille_dict

# Hàm hiển thị chữ Braille
def display_braille(dots_list):
    braille_lines = ["", "", ""]
    for dots in dots_list:
        braille_template = [
            ['○', '○'],
            ['○', '○'],
            ['○', '○']
        ]
        
        coordinates = {
            '1': (0, 0),
            '2': (1, 0),
            '3': (2, 0),
            '4': (0, 1),
            '5': (1, 1),
            '6': (2, 1)
        }
        
        for dot in dots:
            if dot in coordinates:
                x, y = coordinates[dot]
                braille_template[x][y] = '●'
        
        for i in range(3):
            braille_lines[i] += ' '.join(braille_template[i]) + "  "
    
    return braille_lines

# Hàm dịch câu sang chữ nổi
def translate_sentence_to_braille(sentence, braille_dict):
    braille_output = ["", "", ""]
    
    sentence_split = sentence.split()
    
    for idx, word in enumerate(sentence_split):
        if word in braille_dict:  # Kiểm tra nếu từ đặc biệt có trong từ điển Braille
            braille_char = display_braille(braille_dict[word])
            for i in range(3):
                braille_output[i] += braille_char[i]
        else:  # Xử lý từng ký tự trong từ
            for char in word:
                if char in braille_dict:
                    braille_char = display_braille(braille_dict[char])
                    for i in range(3):
                        braille_output[i] += braille_char[i]
                else: 
                    print("Không tìm thấy chữ: " + char)

        # Thêm khoảng cách giữa các từ trừ từ cuối cùng
        if idx < len(sentence_split) - 1:
            a = display_braille(braille_dict.get(" ", ["0"]))
            for i in range(3):
                braille_output[i] += a[i]
                
    print("Câu nhập vào: " + sentence)
    for line in braille_output:
        print(line)

# Đọc dữ liệu từ file CSV
braille_dict = read_braille_data('braille_codes.csv')

# Nhập câu cần dịch
input_sentence = "hi"
translate_sentence_to_braille(input_sentence, braille_dict)
```
